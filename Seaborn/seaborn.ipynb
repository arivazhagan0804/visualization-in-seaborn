## Matrix plot

```python
pivot=df.pivot_table(values="passengers",index='month',columns='year')
pivot
```

## headmap
headmap creates matrix of a data which is already in the matrix format

```python
fig,axes=plt.subplots(figsize=(10,6))
sns.heatmap(pivot,linewidths=0.2,annot=True,linewidths=2)
```

```python
ax,fig=plt.subplots(figsize=(10,6),dpi=150)
sns.heatmap(data=pivot.loc[:,1957:],annot=True,linewidths=0.2)
```

## Grid plot

      grid plot create a plot multiplot in the  rows and columns.

## pairgrid

    It provides possible to together multiple plots.A more flexible version of pairplot

```python
iris=sns.load_dataset('iris')
iris
```

```python
iris['species'].unique()
```

```python
sns.pairplot(data=iris,hue=)
```

```python
iris=sns.PairGrid(data=iris)
iris.map_diag(sns.displot)
iris.map_lower(sns.kdeplot)
iris.map_upper(plt.scatter)
```

## FaceGrid
    The most important differnts of facegrid to compaired to pairplot and pairgrid is that we can select the column to visualize.

```python
df=sns.load_dataset('tips')
df
```

```python
FaceGrid=sns.FacetGrid(data=df,row='smoker',col='time')
FaceGrid.map(sns.displot,'tip',kde=True)
```
