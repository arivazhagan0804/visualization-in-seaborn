## Grid plots 

    It allow us to plot multiplots in columns and rows.

## Pair grid
    It provides the posibility to plot together multiplots.It is more flexcible version of the  pair plot.

```python
df=sns.load_dataset('iris')
df
```

```python
sns.pairplot(df)
```

```python
a=sns.PairGrid(df)
a.map_diag(sns.distplot)
a.map_upper(plt.scatter)
a.map_lower(sns.kdeplot)
```

```python
df
```

```python
b=sns.PairGrid(df,hue='species')
b.map_diag(sns.displot)
b.map_upper(plt.scatter)##
b.map_lower(sns.kdeplot)
```

## Facet Grid
  The most important difference of pairplot compared to pairgrid is that we can select columns to visulize.

```python
tips
```

```python
f=sns.FacetGrid(tips,row='smoker',col='time')
f.map(sns.displot,'total_bill',kde=True)
```

```python
f=sns.FacetGrid(tips,row='sex',col='time')
f.map(plt.hist,'total_bill')
```

```python
g=sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
g.map(plt.scatter,'total_bill','tip')
plt.legend()
```
