 ## I will working with a famous titanic dataset for these excercise.For now ,I will focus on the visualization of the data with seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.info()
```

## jointplot
   It give us combines fare and age for bivariate data in dataset

```python
df['age'].mean()
```

```python
sns.set_context('poster',font_scale=.5)
sns.jointplot(x='fare',y='age',data=df,hue='sex')
```

## displot
 It combination of kde(kernal density estimator) and barplot.kde show us probability distribution of data.

 It  will give us maximum count for fare in the data.

```python
sns.displot(df['fare'],bins=10,color='red')
```

## Box plot
 It will give us to identify the outliers in you data.using a method that is a function of the inter quartile range.


```python
plt.figure(figsize=(10,6))
sns.set_context('poster',font_scale=1)
sns.boxplot(x='class',y='age',data=df,palette='rainbow')
```

## swarmplot
 it will give us univarible in our data.

```python
sns.swarmplot(x='class',y='age',data=df,palette='rainbow')
```

## countplot
 It will give us count of sex column in your data.


```python
sns.countplot(data=df,x='sex',palette='rainbow')
```

## headmap
 it will give us corr() relation in your data

```python
df.head(1)
```

```python
df[['fare','age']].corr()
```

```python
sns.heatmap(df[['fare','age']].corr(),linewidths=1,linecolor='yellow')
```

## FacetGrid
   It will give us select the column to visualize.

```python
plt.figure(figsize=(15,6))
sns.set_context('poster',font_scale=.5)
sex=sns.FacetGrid(df,col='sex')
sex.map(plt.hist,'age')
```
