```python
df['days']=pd.to_datetime(df['Date'])
df['year']=df['days'].apply(lambda x:x.year)
df['year'].unique()
```

```python
df['month']=df['days'].apply(lambda x:x.month)
df['month'].unique()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(8,6),dpi=200)
df.groupby(df['days'])['Boxes Shipped'].mean().plot()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(30,6),dpi=200)
month=df.groupby(df['month']).count()
month.plot()
```
